### 회귀모델(점수예측)

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv('./test.csv',encoding='euc-kr')
df
```

```python
for i in range(len(df)):
    df.loc[i,'취득점수'] = max(df.loc[i,'toeic'],df.loc[i,'teps'],df.loc[i,'toeic_sp'],df.loc[i,'opic'])
df
```

```python
for i in range(len(df)):
    if df.loc[i,'toeic'] == df.loc[i,'취득점수']:
        df.loc[i,'적합시험'] = 'gtoeic'
    if df.loc[i,'teps'] == df.loc[i,'취득점수']:
        df.loc[i,'적합시험'] = 'gteps'
    if df.loc[i,'toeic_sp'] == df.loc[i,'취득점수']:
        df.loc[i,'적합시험'] = 'gtoeic_sp'
    if df.loc[i,'opic'] == df.loc[i,'취득점수']:
        df.loc[i,'적합시험'] = 'gopic'
df
```

```python
get_mbti = pd.get_dummies(df['mbti'])
get_male = pd.get_dummies(df['gender'])
get_st_method = pd.get_dummies(df['st_method'])
get_major = pd.get_dummies(df['major'])
get_test = pd.get_dummies(df['적합시험'])
df = pd.concat([df,get_mbti,get_male,get_st_method,get_major,get_test],axis=1)
df = df.drop(['idx','sucss','mbti','gender','st_method','major','적합시험','toeic','teps','toeic_sp','opic'],axis=1)
df
```

```python
x_df = df.iloc[:,:-4]
x_df = x_df.drop(['취득점수'],axis=1)
y_df = df['취득점수']
# x_tr = x_df.iloc[:201]
# y_tr = y_df.iloc[:201]
```

```python
print(x_df.columns)
x_df
```

```python
pgender = '남'
x_df[pgender]
```

```python
# train데이터 split
from sklearn.model_selection import train_test_split
x_tr, x_val,y_tr, y_val = train_test_split(x_df,y_df, test_size = 0.2, random_state = 2021)
```

```python
x_val.columns
```

```python
# minmax 스케일러
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_tr)
x_train_scaled = scaler.transform(x_tr)
x_test_scaled = scaler.transform(x_val)
```

```python
# 랜덤포레스트 회귀
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train_scaled,y_tr)
pred = model.predict(x_test_scaled)
# model.score(pred, y_val)
```

```python
len(pred)
```

```python
# RMSE
from sklearn.metrics import mean_squared_error
def RMSE(y_val,pred):
    return np.sqrt(mean_squared_error(y_val,pred))
print('RMSE : ' + str(RMSE(y_val,pred)))
```

```python
x_val = x_val.reset_index(drop=True)
y_val = pd.DataFrame(y_val).reset_index(drop=True)
pred = pd.DataFrame({'예측값':pred}).reset_index(drop=True)
val = pd.concat([x_val,y_val,pred],axis=1)
val
```

### 분류분석

```python
c_x_df = df.iloc[:,:-4]
c_x_df = c_x_df.drop(['취득점수'],axis=1)
c1_y_df = df['gtoeic']
c2_y_df = df['gteps']
c3_y_df = df['gtoeic_sp']
c4_y_df = df['gopic']
```

```python
c_x_df
```

```python
# train데이터 split
c_x_tr = c_x_df.iloc[:201]
c_x_test = c_x_df.iloc[201:]

c1_y_tr = c1_y_df.iloc[:201]
c1_y_test = c1_y_df.iloc[201:]

c2_y_tr = c2_y_df.iloc[:201]
c2_y_test = c2_y_df.iloc[201:]

c3_y_tr = c3_y_df.iloc[:201]
c3_y_test = c3_y_df.iloc[201:]

c4_y_tr = c4_y_df.iloc[:201]
c4_y_test = c4_y_df.iloc[201:]
```

```python
# minmax 스케일러
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(c_x_tr)
x_train_scaled = scaler.transform(c_x_tr)
x_test_scaled = scaler.transform(c_x_test)
```

```python
# 랜덤포레스트분류
from sklearn.ensemble import RandomForestClassifier
c1model = RandomForestClassifier()
c1model.fit(x_train_scaled,c1_y_tr)

c2model = RandomForestClassifier()
c2model.fit(x_train_scaled,c2_y_tr)

c3model = RandomForestClassifier()
c3model.fit(x_train_scaled,c3_y_tr)

c4model = RandomForestClassifier()
c4model.fit(x_train_scaled,c4_y_tr)

pred1 = c1model.predict(x_test_scaled)
pred2 = c2model.predict(x_test_scaled)
pred3 = c3model.predict(x_test_scaled)
pred4 = c4model.predict(x_test_scaled)
```

```python
c_x_test = c_x_test.reset_index(drop=True)
c1_y_test = pd.DataFrame(c1_y_test).reset_index(drop=True)
c2_y_test = pd.DataFrame(c2_y_test).reset_index(drop=True)
c3_y_test = pd.DataFrame(c3_y_test).reset_index(drop=True)
c4_y_test = pd.DataFrame(c4_y_test).reset_index(drop=True)

pred1 = pd.DataFrame({'토익예측값':pred1}).reset_index(drop=True)
pred2 = pd.DataFrame({'텝스예측값':pred2}).reset_index(drop=True)
pred3 = pd.DataFrame({'토스예측값':pred3}).reset_index(drop=True)
pred4 = pd.DataFrame({'오픽예측값':pred4}).reset_index(drop=True)

classy = pd.concat([c_x_test,c1_y_test,pred1,c2_y_test,pred2,c3_y_test,pred3,c4_y_test,pred4],axis=1)
cl_pred = classy[['토익예측값','텝스예측값','토스예측값','오픽예측값']]
```

```python
for i in range(len(cl_pred)):
    if cl_pred.loc[i,'토익예측값'] == 1 :
        cl_pred.loc[i,'예측적합시험'] = 'TOEIC'
    if cl_pred.loc[i,'텝스예측값'] == 1 :
        cl_pred.loc[i,'예측적합시험'] = 'TEPS'
    if cl_pred.loc[i,'토스예측값'] == 1 :
        cl_pred.loc[i,'예측적합시험'] = 'TOEIC_SP'
    if cl_pred.loc[i,'오픽예측값'] == 1 :
        cl_pred.loc[i,'예측적합시험'] = 'OPIC'
cl_pred['예측적합시험']=cl_pred['예측적합시험'].fillna('ANYTHING')
cl_pred
```

```python
classy['예측적합시험'] = cl_pred['예측적합시험']
classy
```

```python
# ROC_AUC_SCORE
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
print(roc_auc_score(c1_y_test,pred1))
print(roc_auc_score(c2_y_test,pred2))
print(roc_auc_score(c3_y_test,pred3))
print(roc_auc_score(c4_y_test,pred4))
```

```python
print('정확도 : {:.2f}'.format(accuracy_score(c1_y_test,pred1)))
print('정밀도 : {:.2f}'.format(precision_score(c1_y_test,pred1)))
print('재현율 : {:.2f}'.format(recall_score(c1_y_test,pred1)))
print('f1_score : {:.2f}'.format(f1_score(c1_y_test,pred1)))
print('======================================================')
print('정확도 : {:.2f}'.format(accuracy_score(c2_y_test,pred2)))
print('정밀도 : {:.2f}'.format(precision_score(c2_y_test,pred2)))
print('재현율 : {:.2f}'.format(recall_score(c2_y_test,pred2)))
print('f1_score : {:.2f}'.format(f1_score(c2_y_test,pred2)))
print('======================================================')
print('정확도 : {:.2f}'.format(accuracy_score(c3_y_test,pred3)))
print('정밀도 : {:.2f}'.format(precision_score(c3_y_test,pred3)))
print('재현율 : {:.2f}'.format(recall_score(c3_y_test,pred3)))
print('f1_score : {:.2f}'.format(f1_score(c3_y_test,pred3)))
print('======================================================')
print('정확도 : {:.2f}'.format(accuracy_score(c4_y_test,pred4)))
print('정밀도 : {:.2f}'.format(precision_score(c4_y_test,pred4)))
print('재현율 : {:.2f}'.format(recall_score(c4_y_test,pred4)))
print('f1_score : {:.2f}'.format(f1_score(c4_y_test,pred4)))
print('======================================================')
```
